```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1
b = 2
N = 2
x = np.linspace(0, a, N+1)
y = np.linspace(0, b, N+1)
print(x)
print(y)

X, Y = np.meshgrid(x, y)
print(X)
print(Y)
```

```python
f = np.sin(X) * np.cos(Y)
print(f)
plt.contourf(X, Y, f)
plt.colorbar()
```

```python
a = 1
b = 2
N = 2
x = np.linspace(0, a, N+1)
y = np.linspace(0, b, N+1)
z = np.linspace(0, b, N+1)
print(x)
print(y)
print(z)


X, Y, Z = np.meshgrid(x, y, z)
print(X)
print(Y)
print(Z)
```
